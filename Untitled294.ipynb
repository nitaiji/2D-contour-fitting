```python
# def twoD_Gaussian(x, y, amplitude, xo, yo, sigma_x, sigma_y, theta, offset):
#     xo = float(xo)
#     yo = float(yo)    
#     a = (np.cos(theta)**2)/(2*sigma_x**2) + (np.sin(theta)**2)/(2*sigma_y**2)
#     b = -(np.sin(2*theta))/(4*sigma_x**2) + (np.sin(2*theta))/(4*sigma_y**2)
#     c = (np.sin(theta)**2)/(2*sigma_x**2) + (np.cos(theta)**2)/(2*sigma_y**2)
#     g = offset + amplitude*np.exp( - (a*((x-xo)**2) + 2*b*(x-xo)*(y-yo) 
#                             + c*((y-yo)**2)))
    
def gauss_func(x, y, theta):
    a, x0, y0, sigma_x, sigma_y, b, offset = theta[0], theta[1], theta[2], theta[3], theta[4], theta[5], theta[6]
    p = offset + a*np.exp(-((((x-x0)**2)/sigma_x) + (((y-y0)**2)/sigma_y) + 2*b*(x-x0)*(y-y0)))
    return p
```

```python
x = np.linspace(0, 200, 201)
y = np.linspace(0, 200, 201)
X, Y = np.meshgrid(x, y)
```

```python
data = gauss_func(X, Y, [50, 100, 100, 500, 500, 0.0001, 0])
data_noisy = data + 0.2*np.random.normal(size=data.shape)
Z = data_noisy
```

```python
Z.shape
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_surface(X, Y, Z, cmap='plasma')
ax.set_zlim(0,np.max(Z)+2)
plt.show()
```

```python
def fun(theta):
    return (gauss_func(X, Y, theta) - data_noisy).ravel()
```

```python
# fun().shape
```

```python
from scipy.optimize import least_squares
```

```python
theta0 = [50, 100, 100, 500, 500, 0.0001, 0]
```

```python
res3 = least_squares(fun, theta0)
```

```python
res3.x
```

```python
fit = gauss_func(X, Y, res3.x)
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_surface(X, Y, Z)
ax.plot_surface(X, Y, fit, cmap='plasma')
ax.set_zlim(0,np.max(fit)+2)
plt.show()
```

```python
img1 = plt.contourf(X,Y,Z,50, cmap='jet')
img2 = plt.contourf(X,Y,fit,50, cmap='jet')
plt.colorbar(img1)
plt.colorbar(img2)
```
